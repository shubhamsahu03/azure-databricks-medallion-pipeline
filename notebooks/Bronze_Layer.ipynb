```python
dbutils.widgets.text("file_name", "")
p_file_name=dbutils.widgets.get("file_name")

```

```python

df=spark.readStream.format("cloudFiles")\
.option("cloudFiles.format", "parquet")\
.option("cloudFiles.schemaLocation", f"abfss://bronze@databricksetlss.dfs.core.windows.net/checkpoint_{p_file_name}")\
.load(f"abfss://source@databricksetlss.dfs.core.windows.net/{p_file_name}")
```

```python
df.writeStream.format("parquet")\
    .outputMode("append")\
    .option("checkpointLocation", f"abfss://bronze@databricksetlss.dfs.core.windows.net/checkpoint_{p_file_name}")\
    .option("path", f"abfss://bronze@databricksetlss.dfs.core.windows.net/{p_file_name}")\
    .trigger(once=True)\
    .start()
```
